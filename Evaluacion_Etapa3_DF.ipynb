```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl 
```

# **ESPAÑA**

```python
DF = pd.read_csv('DF_clean.csv')
```

```python
#Verificamos información del DataFrame
DF.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
DF.head()
```

```python
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(DF)
```

## **a) host_response_time**

```python
table_a= freq_tbl(DF['host_response_time'])
table_a
```

```python
Filtro_a= table_a[table_a['frequency']>500]
Filtro_a
```

```python
Filtro_a_index= Filtro_a.set_index('host_response_time')
Filtro_a_index
```

```python
Filtro_a_index.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('host_response_time')
plt.xlabel('time')
plt.ylabel('Frecuencia')
```

```python
Filtro_a_index["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f %%")
```
